```python
import billboard
```

```python
charts = billboard.ChartData('hot-100', date="1990-02-01", fetch=True, timeout=25)
# print(charts)
```

```python
from time import sleep
```

```python
i = 1
charts1990 = []
while i <= 4:

#   if i < 10:
#     date1990 = "1990-0" + str(i) +"-01"
#   else:
#     date1990 = "1990-" + str(i) +"-01"
  # print(i)
  # print(date1990)

  charts = billboard.ChartData('hot-100', date="1990-0" + str(i) +"-01", fetch=True, timeout=25)
  # print(charts)
  charts1990.append(charts)
  print(f"1990-0{i}-01 chart appended.")

  i += 1
  sleep(60)
```

```python
# charts1990 = [billboard.ChartData('hot-100', date="1990-01-01", fetch=True, timeout=25)]
```

```python
charts1990
```

```python
song_titles = []

for chart in charts1990:
    count = 0
    while count < len(chart):
        song_titles.append(chart[count].title)
        count += 1
   
# song_titles
```

## Spotify Audio Features

```python
import spotipy
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy.oauth2 as oauth2
```

```python
market = [ "AD", "AR", "AT", "AU", "BE", "BG", "BO", "BR", "CA", "CH", "CL", "CO", "CR", "CY", 
      "CZ", "DE", "DK", "DO", "EC", "EE", "ES", "FI", "FR", "GB", "GR", "GT", "HK", "HN", "HU", 
      "ID", "IE", "IS", "IT", "JP", "LI", "LT", "LU", "LV", "MC", "MT", "MX", "MY", "NI", "NL", 
      "NO", "NZ", "PA", "PE", "PH", "PL", "PT", "PY", "SE", "SG", "SK", "SV", "TH", "TR", "TW", 
      "US", "UY", "VN" ]

CLIENT_ID = "36b848c5ddc54db693616309e88a5971"
CLIENT_SECRET = "92e422dec9324e89a8fbcddf74666b1c"

credentials = oauth2.SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET)

token = credentials.get_access_token()
spotify = spotipy.Spotify(auth=token)
```

```python
output = []
for title in song_titles:
  output.append(spotify.search(title, type="track", market=market, limit=1))
# output
```

```python
ids = []
for track in output:
    if len(track['tracks']['items']) >= 1: 
        ids.append(track['tracks']['items'][0]['id'])
# ids
```

```python
ids
```

```python
# first 100 audio_features
features1 = spotify.audio_features(ids[0:100])
```

```python
# 101-200 audio_features
features2 = spotify.audio_features(ids[101:200])
```

```python
# 200-300 audio_features
features3 = spotify.audio_features(ids[201:300])
```

```python
# >300 audio_features
features4 = spotify.audio_features(ids[300:])
```
